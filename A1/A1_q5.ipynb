```python
import numpy as np
import matplotlib.pyplot as plt

size = np.array([1, 1000, 2000, 4000])
time = np.array([2.17e-4, 2.95e-4, 3.03e-4, 8.98e-4]) 

A = np.array([[np.sum(size**2), np.sum(size)],[np.sum(size), len(size)]])
b = np.array([[np.sum(size * time)], [np.sum(time)]])

coefficients = np.linalg.solve(A, b)

print("Coefficients (a, b):", coefficients.flatten())

plt.scatter(size, time, label='Data Points', color='red')
plt.plot(size, coefficients[0] * size + coefficients[1], label='Fitted Curve', color='blue')
plt.xlabel('Size')
plt.ylabel('Time')
plt.legend()
plt.show()
```
