```python
import matplotlib.pyplot as plt
import numpy as np

def plot(x,y,title,xlabel,ylabel,y_titles = None,log_xaxis=False,log_yaxis=False):
    plt.plot(x,y,marker = 'o')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    if log_xaxis:
        plt.xscale('log')
    if log_yaxis:
        plt.yscale('log')
    plt.legend()
    plt.show()
```

```python
# A1 q2a plot
x = np.array([2, 4, 8, 16])
y_allreduce = np.array([0.000001, 0.000007,0.16838,0.057251])
y_globalsum = np.array([0.000004, 0.000003,0.018494, 0.048490])
plot(x, y_allreduce, 'A1 Q2a: All_Reduce timing', 'p', ' Log Time elapsed (sec)', log_yaxis=True)
plot(x, y_globalsum, 'A1 Q2a: Global_Sum timing', 'p', ' Log Time elapsed (sec)', log_yaxis=True)
```
