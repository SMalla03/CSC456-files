```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# A1 q2 plot
x = np.array([2, 4, 8, 16])
y_allreduce = np.array([0.000000529, 0.000004751,0.000005986,0.000009014])
y_globalsum = np.array([0.000000265,0.000003813,0.000003946,0.000005596])
y_MPIreducemaxminprod = np.array([0.000001524,0.000010876,0.000015724,0.000021627])
y_globalreducemaxminprod = np.array([0.000000558,0.000007662,0.000009124,0.000011626])
plt.plot(x,y_allreduce,marker = 'o', label='MPI_Allreduce_Sum',c = 'blue')
plt.plot(x,y_globalsum,marker = 'o', label='Global_Sum',c = 'orange')
plt.plot(x,y_MPIreducemaxminprod,linestyle = '--',marker = 'o', label='MPI_Allreduce_Sum_Max_Min_Prod',c = 'blue')
plt.plot(x,y_globalreducemaxminprod,linestyle = '--',marker = 'o', label='Global_Sum_Max_Min_Prod',c = 'orange')
plt.title('A1 Q2: All_Reduce vs Global algorithm timing')
plt.xlabel('Number of processors')
plt.ylabel('Log Time elapsed (sec)')
plt.yscale('log')
plt.grid(True)
plt.legend(loc='lower right',fontsize = 9)
plt.show()
```
